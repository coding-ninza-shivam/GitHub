```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
```

```python
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)

## plotting with dots (without line)
plt.plot(xpoints, ypoints, 'o')

## marker style
plt.plot(ypoints, marker = '*')

## line style, color, width
plt.plot(xpoints, linestyle = 'dotted', color = 'hotpink', linewidth = '4')

## title and x-y lables
plt.title("X vs Y graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

## grid
plt.grid()
# plt.grid(axis = 'x') for grid in x axis only
# can also set line propertise for grid
# plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

## show plot
plt.show()
```

```python
## Drawing multiple plots

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
# plt.subplot(nrows, ncols, index)
# nrows=1: 1 row
# ncols=2: 2 columns
# index=1: This refers to the 1st subplot (i.e., the left one)

# So plt.subplot(1, 2, 1) means:
# "Create a figure with 1 row and 2 columns of plots, and activate the 1st plot (left plot) for drawing."
# plt.plot(x,y)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)


plt.show()
```

```python
## Bar plot ##
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, color= "#4CAF50", width = 0.7) #by default width is 0.8
# plt.barh(x, y) for horizontal bar plot
# height = 0.6 (for horizontal bar use height in spite of width)

plt.show()
```

```python
## Histogram ##

x = np.random.normal(170, 10, 250) # creating histogram

plt.hist(x)
plt.show()
```

```python
## Pie chart ##

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
# By default the plotting of the first wedge starts from the x-axis and moves counterclockwise

# Lables, Explode, Color and Shadow

mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0, 0, 0, 0.3]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, colors = mycolors, shadow = True )

## Legend 
plt.legend() #can also be used with header plt.legend(title = "Four Fruits:")
plt.show() 
```
